### This is an overview of the 2020 electoral votes for Ghana.
- It starts off with analyzing the total votes cast across all 16 regions.
- Then it analyzes votes by region for each political party.
- Focus is now given to the 2 major political parties in Ghana - NDC and NPP.
    - total votes by region for both NDC and NPP is analyzed.
    - regions where NDC had majority of the total votes cast is analyzed.
    - regions where NPP had majority of the total votes cast is analyzed.

#### We'll start by importing the necessary libraries.

```python
# import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Load and access the dataset

```python
# Load the Excel file
file_path = "ghana_2020_elections_data.xlsx"
excel_file = pd.ExcelFile(file_path)

# Get sheet names
sheet_names = excel_file.sheet_names
sheet_names
```

- The votes were cast in the sixteen (16) regions of Ghana.

- There are 16 sheets in the workbook, each sheet representing the votes cast in a region. 

```python
# Load one of the sheets
df = pd.read_excel(file_path, sheet_name='ahafo')
df.head()
```

```python
df.columns
```

All of the 16 sheets do not have a `region` column to identify the region the votes were cast apart from the sheet names.

#### Add the `region` column to each sheet and save to a new Excel file.

```python
new_file_path = "ghana_2020_elections_data_modified.xlsx"

# Function to capitalize each word in a string (sentence case)
def capitalize_each_word(string):
    return ' '.join(word.capitalize() for word in string.replace('_', ' ').split())

# Create a dictionary to hold DataFrames
df_dict = {}

# Read each sheet, add the 'region' column, and store it in the dictionary
for sheet in sheet_names:
   df = pd.read_excel(file_path, sheet_name=sheet)
   region_value = f'{sheet} Region'
   df['region'] = capitalize_each_word(region_value)
   df_dict[sheet] = df

# Save each modified DataFrame back to a new Excel file
with pd.ExcelWriter(new_file_path, engine='openpyxl') as writer:
   for sheet, df in df_dict.items():
       df.to_excel(writer, sheet_name=sheet, index=False)

print("Columns added and saved to the new file successfully!")
```

```python
# Check if columns have been added
df = pd.read_excel(new_file_path, sheet_name="central")
df.head()
```

#### Merge the sheets in the workbook 

```python
# Read each sheet into a DataFrame and concatenate them
df_merged = pd.concat([pd.read_excel(new_file_path, sheet_name=sheet) for sheet in sheet_names],
                      ignore_index=True)

# Save the merged DataFrame to a new csv file
df_merged.to_csv("merged_ghana_2020_elections_data_modified.csv",
                index=False)
                      
```

```python
# Access the merged file 
df_new = pd.read_csv("merged_ghana_2020_elections_data_modified.csv")
df_new
```

#### Analyzing the total votes cast across all 16 regions

```python
# Group the data by region and sum the votes
region_votes = df_new.groupby('region')['votes'].sum().reset_index().sort_values(by='votes', ascending=False)
region_votes
```

```python
# Plot the data using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='votes', y='region', data=region_votes, hue='region', legend=False)
plt.title('Total Votes Cast by Region')
plt.xlabel('Votes')
plt.ylabel('Region')
# plt.savefig('Total Votes Cast by Region.png')
plt.show()
```

- The Greater Accra Region had the highest number of votes cast, followed by the Ashanti Region, even though the former has a smaller land mass compared to the latter.
- The Eastern Region had the third highest, followed by the Central Region all in excess of 1,000,000 votes.
- All the other regions had less than 1,000,000 votes cast in each region with Savannah Region having the lowest votes cast - 229, 085.

#### Analyzing the distribution of votes for the various political parties.

```python
# Pivot the data to get votes by region and party
pivot_df = df_new.pivot_table(index='region', columns='party_initials', values='votes', aggfunc='sum').fillna(0)
pivot_df
```

```python
# Create a new column for the total votes in each region
pivot_df['Total_Votes'] = pivot_df.sum(axis=1)

# Sort the pivot table by total votes in descending order
pivot_df = pivot_df.sort_values(by='Total_Votes', ascending=True).drop(columns='Total_Votes')

# Plot the data
pivot_df.plot(kind='barh', stacked=True, figsize=(12, 8), colormap='viridis')
plt.title('Votes by Region for Each Political Party')
plt.xlabel('Region')
plt.ylabel('Votes')
plt.legend(title='Party Initials')
plt.tight_layout()
# plt.savefig('Votes by Region for Each Political Party.png') #uncommennt to save plot
plt.show()
```

#### Focusing on the 2 major political parties in Ghana - NDC and NPP

```python
# Categorize other parties as 'Other'
df_new = df_new.copy()
df_new['party_initials'] = df_new['party_initials'].replace({'NPP': 'NPP', 'NDC': 'NDC'}, regex=True)
df_new['party_initials'] = df_new['party_initials'].apply(lambda x: x if x in ['NPP', 'NDC'] else 'Other')
df_new['full_party_name'] = df_new['full_party_name'].apply(lambda x: x if x in ['New Patriotic Party', 'National Democratic Congress'] else 'Other')

df_new.head()
```

Let's create a new DataFrame that contains data for the two major political parties.

```python
major_parties = ['NDC', 'NPP']
major_parties_df = df_new[df_new['party_initials'].isin(major_parties)]
major_parties_df.head()
```

#### Analyzing total votes by region for both NDC and NPP 

```python
# Group the data by region and sum the votes for the 2 parties
region_votes_mp = major_parties_df.groupby('region')['votes'].sum().reset_index().sort_values(by='votes', ascending=False)
region_votes_mp
```

```python
# Plot the data using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='votes', y='region', data=region_votes_mp, hue='region', legend=False)
plt.title('Total Votes by Region for NDC and NPP')
plt.xlabel('Votes')
plt.ylabel('Region')
# plt.savefig('Total Votes by Region for NDC and NPP.png') #uncommennt to save plot
plt.show()
```

```python
# Pivot the data to get votes by region and party
pivot_df_mp = major_parties_df.pivot_table(index='region', columns='party_initials', values='votes', aggfunc='sum').fillna(0)
pivot_df_mp
```

```python
# Create a new column for the total votes in each region
pivot_df_mp['Total_Votes'] = pivot_df_mp.sum(axis=1)

# Sort the pivot table by total votes in descending order
pivot_df_mp = pivot_df_mp.sort_values(by='Total_Votes', ascending=True).drop(columns='Total_Votes')

# Plot the data
pivot_df_mp.plot(kind='barh', stacked=True, figsize=(12, 8), color=['green', 'blue', 'grey'])
plt.title('Votes by Region for NDC and NPP')
plt.xlabel('Region')
plt.ylabel('Votes')
plt.legend(title='Party Initials')
plt.tight_layout()
# plt.savefig('Votes by Region for NDC and NPP.png') #uncommennt to save plot
plt.show()
```

#### Comparing the total votes cast for all parties vs the two major parties (NPP and NDC)

```python
# Define the order based on the pivot table's regions
region_order = pivot_df.index.tolist()

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 12), sharey=True, sharex=True)

# First Plot: Total Votes Cast by Region
sns.barplot(ax=axes[0, 0], x='votes', y='region', data=region_votes, hue='region', dodge=False, order=region_order, legend=False)
axes[0, 0].set_title('Total Votes Cast by Region')
axes[0, 0].set_xlabel('Votes')
axes[0, 0].set_ylabel('Region')

# Second Plot: Total Votes by Region for NDC and NPP
sns.barplot(ax=axes[1, 0], x='votes', y='region', data=region_votes_mp, hue='region', dodge=False, order=region_order, legend=False)
axes[1, 0].set_title('Total Votes by Region for NDC and NPP')
axes[1, 0].set_xlabel('Votes')
axes[1, 0].set_ylabel('')

# Third Plot: Votes by Region for Each Political Party 
pivot_df['Total_Votes'] = pivot_df.sum(axis=1)
pivot_df = pivot_df.sort_values(by='Total_Votes', ascending=True).drop(columns='Total_Votes')
pivot_df.plot(kind='barh', stacked=True, ax=axes[0, 1], colormap='viridis')
axes[0, 1].set_title('Votes by Region for Each Political Party')
axes[0, 1].set_xlabel('Votes')
axes[0, 1].set_ylabel('Region')

# Fourth Plot: Votes by Region for NDC and NPP 
pivot_df_mp['Total_Votes'] = pivot_df_mp.sum(axis=1)
pivot_df_mp = pivot_df_mp.sort_values(by='Total_Votes', ascending=True).drop(columns='Total_Votes')
pivot_df_mp.plot(kind='barh', stacked=True, ax=axes[1, 1], color=['green', 'blue'])
axes[1, 1].set_title('Votes by Region for NDC and NPP')
axes[1, 1].set_xlabel('Votes')
axes[1, 1].set_ylabel('')

plt.tight_layout()
# plt.savefig('Total Votes Cast for All Parties Vs the Two Major Parties (NPP and NDC).png') #uncommennt to save plot
plt.show()
```

From the above plot, it can be concluded that:
- the order of the votes by region remains the same even after focusing on the two major parties.
- the votes of the other parties are insignificant compared to the two major parties. 


#### Analyzing the votes cast for the two major parties - NDC and NPP

For every region, between the two major parties:
- determine the party that had majority of the votes.
- count the total majority votes by region won for each party.
- determine the total votes for the regions won.
- plot the data for both parties.

```python
# Function to find the party with the majority votes in each region
def majority_party_by_region(df):
    majority_df = df.loc[df.groupby('region')['votes'].idxmax()]
    return majority_df

# Get the majority party for each region within the filtered DataFrame
majority_df = majority_party_by_region(major_parties_df).sort_values(by='votes', ascending=False)
majority_df
```

```python
# Count the number of regions won by each party
majority_df.groupby('party_initials')['region'].count()
```

```python
# Sum the total votes for the regions won by NPP and NDC
mp_votes_total = majority_df.groupby('party_initials')['votes'].sum()
mp_votes_total
```

```python
# Plot the data for the regions won by each party
parties = ['NDC', 'NPP']
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 6), sharey=True)

# Define the order of regions to maintain consistent y-axis order
region_order = majority_df['region'].unique().tolist()

for idx, party in enumerate(parties):
    party_df = majority_df[majority_df['party_initials'] == party]
    sns.barplot(ax=axes[idx], x='votes', y='region', data=party_df, hue='region', dodge=False, order=region_order, legend=False)
    axes[idx].set_title(f'Regions where {party} had the Majority Votes')
    axes[idx].set_xlabel('Votes')
    axes[idx].set_ylabel('Region' if idx == 0 else '')  # Only set y-label on the first plot

# Adjust layout
plt.tight_layout()
# plt.savefig('Majority Votes by Region for NPP and NDC.png') #uncommennt to save plot
plt.show()
```

```python
# Plot the percentages of the majority votes by region for NDC and NPP
plt.pie(mp_votes_total, colors=['green', 'blue'], 
        explode=[0, 0.01], labels=['NDC', 'NPP'], autopct='%.0f%%')
plt.title("Percentage win of Majority Votes for NDC and NPP")
# plt.savefig('Percentage Win of Majority Votes for NDC and NPP') #uncommennt to save plot
plt.show()
```

#### Pecentage of total votes cast 

```python
# Calculate total votes cast in all the regions 
total_votes = df_new.groupby(by='party_initials')['votes'].sum() 
total_votes
```

```python
# Plot the percentages of the total votes cast for NDC and NPP
plt.pie(total_votes, colors=['green', 'blue', 'orange'], 
        explode=[0.01, 0.01, 0.00], labels=['NDC', 'NPP', 'Other'], autopct='%.2f%%')
plt.title("Percentage win of Total Votes")
# plt.savefig('Percentage win of Total Votes.png') #uncommennt to save plot
plt.show()
```

#### Conclusion

- Even though the NDC won the majority of the regions (9 out of 16), the NPP had the highest total majority votes from 7 regions.
- The margin of votes won by the NPP in its majority regions was significantly larger compared to that of the NDC.
- The NPP received the majority of its votes from the southern part of Ghana, whereas the NDC had their majority votes from the northern part of Ghana.
- Out of the total votes cast, the NPP had the majority with 51.30%, followed by the NDC with 47.36%, and the other political parties with 1.34%.
